```python
import pandas as pd
import numpy as np
from faker import Faker

fake = Faker()

customer_ids = []
ages = []
genders = []
categories = []
amounts = []
dates = []

for i in range(1000):
    customer_ids.append(fake.uuid4())
    ages.append(np.random.randint(18, 60))
    genders.append(np.random.choice(["Male", "Female"]))
    categories.append(np.random.choice(["Electronics", "Clothing", "Groceries"]))
    amounts.append(round(np.random.uniform(10, 500), 2))
    dates.append(fake.date_this_year())

df = pd.DataFrame({
    "Customer_ID": customer_ids,
    "Age": ages,
    "Gender": genders,
    "Product_Category": categories,
    "Purchase_Amount": amounts,
    "Purchase_Date": dates
})

df.to_csv("synthetic_retail_data.csv", index=False)

print("Synthetic dataset generated successfully")
```
